```python
import pandas as pd
import fastf1
from fastf1 import plotting

year, gp, ses = 2024, "Monaco", "R"  # Race
session = fastf1.get_session(year, gp, ses)
session.load(laps=False, telemetry=False, weather=False)

res = session.results  # SessionResults -> pandas.DataFrame
print(res.columns)
display(res.head(10))

# Handy columns: Position (finish), GridPosition (start), Abbreviation/FullName, TeamName, Points, Status, Laps, etc.
# (Columns are standardized even if some are N/A depending on session type.)
```
